# **Missing values check**

categorical columns fill with *mode*

```python
cat_cols= df.select_dtypes(include="object").columns
for col in cat_cols:
  df[col]=df[col].fillna(df[col].mode()[0])
```

```python
df.isnull().sum().sort_values(ascending= False)
```

```python
df['Alley'].fillna(df['Alley'].mode()[0],inplace=True)
df['MasVnrType'].fillna(df['MasVnrType'].mode()[0],inplace=True)
df['Electrical'].fillna(df['Electrical'].mode()[0],
inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum().sort_values(ascending=False).head(10)
```

```python
if'Unnamed:0'in df.columns:
  df.drop('Unnamed:0',axis=1,inplace=True)
```

## **visualization by graph**


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

```

# saleprice distribution (histogram)

```python
plt.figure(figsize=(8,4))
sns.histplot(df["SalePrice"],kde=True)
plt.title("SalePrice Distribution")
plt.savefig("graph1.png",dpi=300,
bbox_inches="tight")
plt.show()
```

# Analysis 2+ Variable

```python
plt.figure(figsize=(12,8))
corr=df.corr(numeric_only=True)
sns.heatmap(corr,annot=False,
cmap="coolwarm")
plt.title("Coorelation Heatmap")
plt.savefig("graph1.png",dpi=300,
bbox_inches="tight")
plt.show()
```

```python
corr_target =corr['SalePrice'].sort_values(ascending=False)
print(corr_target.head(10))
```

# **Scatterplot: GrlivArea vs SalePrice**

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x=df['GrLivArea'],y=df['SalePrice'])
plt.title('SalePrice vs GrLivArea')
plt.savefig("graph1.png",dpi=300,
bbox_inches="tight")
plt.show()
```

# Boxplot:OverallQual Vs Saleprice

```python
plt.figure(figsize=(8,4))
sns.boxplot(x=df['OverallQual'],y=df['SalePrice'])
plt.title("SalePrice vs OverallQual")
plt.savefig("graph1.png",dpi=300,
bbox_inches="tight")
plt.show()
```

```python
df.columns
```

Top 10 Neighborhoods Avg saleprice (Barplot)

```python
top10= df.groupby("Neighborhood",
 as_index=False)["SalePrice"].mean()
top10= top10.sort_values("SalePrice",
 ascending=False).head(10)

plt.figure(figsize=(12,5))
sns.barplot(data=top10,x="Neighborhood",
y="SalePrice")
plt.xticks(rotation=45)
plt.title("Top 10 Neighborhoods by Avg SalePrice")
plt.savefig("graph1.png",dpi=300,
bbox_inches="tight")
plt.show()
```

```python
missing=df.isnull().sum().sort_values(ascending=False)
missing=missing[missing > 0 ]

plt.figure(figsize=(10,4))
missing.head(15).plot(kind='bar')
plt.title("Top Missing Value Columns")
plt.savefig("graph1.png",dpi=300,
bbox_inches="tight")
plt.show()
```

```python
df['HouseAge']=df["YrSold"]-df["YearBuilt"]
df["TOtalBathrooms"]=df["FullBath"]
0.5*df["HalfBath"]+ df["BsmtFullBath"]
0.5*df["BsmtHalfBath"]
df["TotalSF"]=df["TotalBsmtSF"] + df["1stFlrSF"]
df["2ndFlrSF"]
```

```python
trend=df.groupby("YrSold")["SalePrice"].mean()

plt.figure(figsize=(7,4))
trend.plot(kind="line",marker="1")
plt.title("Avg SalePrice Trend by Year Sold")
plt.ylabel("Avg SalePrice")
plt.grid(True)
plt.savefig("SalePrice_hist.png",dpi=300,
bbox_inches="tight")
plt.show()
```
